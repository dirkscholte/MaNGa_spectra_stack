# Exploring the FastSpecFit results

```python
import glob

from astropy.io import fits
from astropy.table import Table

import matplotlib.pyplot as plt
```

```python
spectra_files = glob.glob('Spectra*.csv')
```

```python
i=1
spectra_csv = Table.read(spectra_files[i])
spectra_fits = fits.open(spectra_files[i][:-3]+'fits')
fastspec_table = Table.read('stackfit_'+spectra_files[i][:-3] +'fits')
```

```python
spectra_files[i]
```

```python
cmap = plt.get_cmap('cividis_r')
for i in range(len(spectra_fits['FLUX'].data)):
    plt.plot(spectra_fits['WAVE'].data, spectra_fits['FLUX'].data[i], c=cmap(i/len(spectra_fits['FLUX'].data)))
plt.title('yellow = center, black = edge')
```
